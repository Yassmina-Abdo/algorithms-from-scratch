## 1. Draw Graph

```python

G = {    '0': ['1', '3','8'],
         '1': ['0','7'],
         
         '2': ['3','5','7'],
         '3': ['0','2','4'],
         '4': ['3', '8'],
         '5': ['2', '6'],
         '6': [],
         '7': ['1', '2'],
         '8': ['0', '4']
    }

print(G)
```

```python
import networkx as nx
g= nx.Graph(G)
nx.draw(g, with_labels=True)
```

## 2. DFS

```python
list(nx.dfs_preorder_nodes(g))
```

## 3. BFS

```python
list(nx.bfs_tree(g,source='0'))
```

## 4. Dijkstra weighted Graph

```python
gg = nx.Graph()
gg.add_edge(0, 1, weight=8)
gg.add_edge(0, 2, weight=9)
gg.add_edge(1, 3, weight=1)
gg.add_edge(1, 5, weight=7)
gg.add_edge(1, 6, weight=9)
gg.add_edge(2, 5, weight=6)
gg.add_edge(2, 6, weight=4)
gg.add_edge(3, 7, weight=4)
gg.add_edge(5, 7, weight=2)
gg.add_edge(7, 6, weight=6)
gg.add_edge(6, 4, weight=7)
gg.add_edge(4, 7, weight=7)

labels=nx.get_edge_attributes(gg , 'weight' )
pos = nx.spring_layout(gg , seed=8)
nx.draw_networkx_edge_labels(gg , pos,edge_labels=labels ,font_weight='bold')
nx.draw(gg,pos , with_labels=True)
```

```python
p1 = nx.shortest_path(gg, source=0, weight='weight')
print(f"All paths from 0: {p1}")
```
